```python
from sklearn.feature_extraction.text import CountVectorizer

# Sample documents
documents = [
    "This is the first document.",
    "This document is the second document.",
    "And this is the third one.",
    "Is this the first document?",
]

# Preprocess the documents to remove "maida"
documents_without_maida = [doc.replace("maida", "") for doc in documents]

# Create a CountVectorizer instance
vectorizer = CountVectorizer(max_features=5000, stop_words='english')

# Fit the vectorizer to the documents and transform the documents into vectors
X = vectorizer.fit_transform(documents_without_maida)

# Get the vocabulary (unique words)
vocab = vectorizer.get_feature_names_out()

# Print the vocabulary and the vector representation of each document
print("Vocabulary:", vocab)
print("Vector representation of documents:")
print(X.toarray())
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        ps.stem(i)
    return " ".join(y)
```

```python
ps.stem('milked')
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(X)
```

```python
sorted(similarity[0],reverse=True)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(food):
    food_index=new_df[new_df['name']=='food']
    distances=similarity[food_index]
    food_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in food_list: 
        print(i)
    return
```
